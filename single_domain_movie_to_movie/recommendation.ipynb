```python
unique_genres = list(set(data['genres']))
print("unique genres len: ", len(unique_genres))
print("unique genres: ", unique_genres)
```

```python
# Prerana

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Finding Cosine similarity between genre.
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(unique_genres)
similarity_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)
print("tfidf_matrix matrix: ", tfidf_matrix)
print("similarity matrix: ", similarity_matrix)
```

```python
# Prerana

from collections import defaultdict
from operator import itemgetter

#similarity_map key = genre, value = all genres similar to all key genre 
similarity_map = defaultdict(list)
similarity_threshold = 0.85

for genre_index, row in enumerate(similarity_matrix):
    genre = unique_genres[genre_index]                                               # a[0]
    for other_genre_score_index, other_genre_score in enumerate(row):
        if other_genre_score > similarity_threshold:
            other_genre = unique_genres[other_genre_score_index]
            similarity_map[genre].append((other_genre, other_genre_score))           #tuple

for k, v in similarity_map.items():
    print("genre: ", k, "similar genres: ", sorted(v, key=itemgetter(1), reverse=True))
```

```python
def get_recommended_movies(user_id, num=20, verbose=False, print_recommendations=False):
    
    # To fetch all the movies watched by a given user.
    
    filtered_user_movies = data[data["userId"] == user_id]
    if verbose:
        print("Initial filtered_user_movies list ", filtered_user_movies)
    
    # Calculating average rating of the movies rated by a given user.
    
    user_rating_mean = filtered_user_movies["rating"].mean()
    if verbose:
        print("The average rating given by the user:: user_rating_mean ", user_rating_mean)
    
    
    # fetching movies(watched by the user) whose rating is greater than mean rating of the user.
    # This is to show him only those genre movies which he likes (higher rating given by the user). 
    
    filtered_user_movies = filtered_user_movies[filtered_user_movies["rating"] > user_rating_mean]
    
    if verbose:
        print("Movies which have rating greater than mean rating:: filtered_user_movies", filtered_user_movies.head())
    
    
    # fetching unique genre
    
    user_genres = set(filtered_user_movies["genres"])
    if verbose:
        print("user genres : ", user_genres)
        print("user genres length : ", len(user_genres))
    
    
    # To fetch genres similar to the genre liked by the user.
    
    all_similar_genres = []
    for user_genre in user_genres:
        all_similar_genres += similarity_map[user_genre]
    if verbose:
        print("List of all similar genres :: all_similar_genres ", all_similar_genres[:10])
    
    
    # Sorting the similar genres based on cosine similarity.
    
    sorted_genres = sorted(all_similar_genres, key=itemgetter(1), reverse=True)
    if verbose:
        print("sorted_genres ", sorted_genres)
        print("==================================")
    
    
    # Fetching recommended movie titles based on similar genre liked by the user
    
    recommended_movie_titles = []
    running_count_of_titles = 0
    for sorted_genre in sorted_genres:
        # To show movies not watched by the user.
        filtered_movies = data[(data["userId"] != user_id) & (data["genres"] == sorted_genre[0])] 
        filtered_movie_titles = list(filtered_movies["title"].unique())
        
        running_count_of_titles += len(filtered_movie_titles)        
        recommended_movie_titles += filtered_movie_titles
    
        if running_count_of_titles >= num:
            break
    
    # list of movies to be recommended
    if verbose:
        print("recommended_movie_titles: ", recommended_movie_titles)

    # Sorting recommended movies based on it's rating.
    
    recomm_movie_based_on_rating = ratings[ratings["title"].isin(recommended_movie_titles)]
    sorted_recomm_movie_based_on_rating = recomm_movie_based_on_rating.sort_values(by="rating", ascending=False)
    
    # Final top 20 movie recommendation list 
    if print_recommendations:
        print("<<<<=====================================================================>>>>")
        print("Final movie recommendation list for a given user: ", user_id)
        print(sorted_recomm_movie_based_on_rating["title"][:num])
    
    recommended_movies_for_user = sorted_recomm_movie_based_on_rating[:num]
    relevant_recommended_movies_len = len(recommended_movies_for_user[recommended_movies_for_user["rating"] > 3.5])
    total_number_of_recommended_movies_len = len(sorted_recomm_movie_based_on_rating)

    total_relevant_movies = data[(data["userId"] != user_id) & (data["rating"] > 3.5) & data["genres"].isin(user_genres)]
    total_relevant_movies_len = len(total_relevant_movies["title"].unique())
    
    precision = relevant_recommended_movies_len / num
    recall = relevant_recommended_movies_len / total_relevant_movies_len
    return precision, recall

    
```

```python
# Recommendations
get_recommended_movies(29, 10, print_recommendations=True) #recommended movie list for user_id = 29
get_recommended_movies(46, 10, print_recommendations=True) #recommended movie list for user_id = 46
```

```python
# Experimentation
import random
from statistics import mean
for num_movies_to_recommend in [1, 5, 10, 15, 50, 75]:
    all_precisions, all_recalls = [], []
    random_user_ids = random.choices(user_ids, k=10)
    for user_id in random_user_ids:
        precision, recall = get_recommended_movies(user_id, num_movies_to_recommend, verbose=False)
        all_precisions.append(precision)
        all_recalls.append(recall)

    print("num movies to recommend: ", num_movies_to_recommend)
    print("mean precision: ", mean(all_precisions))
    print("mean recall: ", mean(all_recalls))
```
